```python
import saxtal_functions as sax
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import os
```

```python
# Change working directory
os.chdir('/home/rkarimi/saxtal/IY_dataset/motion_corrected')
# Run the loop to rebin all the pbs-apo motion corrected images
for i in tqdm(os.listdir()):
    if i != "masked_output":
        sax.mask_image(i, 
               threshold_method='sd', 
               verbose=True, 
               threads=16, 
               sigma=15, 
               num_sd=2.5, 
               x_window_percent=(0, 0.7),
               y_window_percent=(0, 0.995),
               mask_hotpixels=False,
               replace_distance_percent=0.05,
               return_spots=False)
```
